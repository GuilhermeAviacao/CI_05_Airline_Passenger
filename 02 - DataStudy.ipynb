```python

# Exclusion of flight column
# Only encode categorical columns that make sense for modeling
categorical_for_encoding = ['airline', 'source_city', 'departure_time', 'stops',
                           'arrival_time', 'destination_city', 'class']
```

```python

# Remove 'flight' since it has too many unique values
print(f"\nEncoding only: {categorical_for_encoding}")
```

```python

from feature_engine.encoding import OneHotEncoder
```

```python

encoder = OneHotEncoder(variables=categorical_for_encoding, drop_last=False)
df_ohe = encoder.fit_transform(df)
```

```python

print(f"Original shape: {df.shape}")
print(f"After encoding shape: {df_ohe.shape}")
```

```python
# Check what columns are in the encoded dataframe
print("Columns in encoded dataframe:")
print(df_ohe.columns.tolist())
```

```python

# Remove the flight column since it's still text and can't be used for correlation
df_ohe_numeric = df_ohe.drop('flight', axis=1)
print(f"\nShape after removing flight column: {df_ohe_numeric.shape}")
```

```python

# Calculate correlation with price (note: lowercase 'price')
corr_spearman = df_ohe_numeric.corr(method='spearman')['price'].sort_values(key=abs, ascending=False)
```

```python

# Remove price correlation with itself and show top 10
print("\nTop 10 correlations with price:")
top_correlations = corr_spearman[1:11]  # Skip first one (price with itself)
print(top_correlations)
```

```python

# Also try Pearson correlation
print("\nPearson correlation (top 10):")
corr_pearson = df_ohe_numeric.corr(method='pearson')['price'].sort_values(key=abs, ascending=False)
top_pearson = corr_pearson[1:11]
print(top_pearson)
```

```python
# Create a simple visualization
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
top_correlations.plot(kind='barh')
plt.title('Top 10 Features Correlated with Price (Spearman)')
plt.xlabel('Correlation Coefficient')
plt.tight_layout()
plt.show()
```
